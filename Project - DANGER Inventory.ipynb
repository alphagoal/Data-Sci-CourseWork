## DANGER Inventory


### Danger Restaurant is selling some products: "Cup", "Dish", "Fork", "Spoon"



**Initially, there are 10 Cups, 20 Dishes, 5 Forks and 1 Spoons in the stop.**



1. Create a dictionary call ```inventory```, storing the number of each item. It should be in the format "item_name": item_count.

````python
# Type your code below
inventory={
    'Cup':10,
    'Dish':20,
    'Fork':5,
    'Spoon':1
}

print(inventory)
````

2. Create a function call ```available_item()```, printing out the inventory with the amount remaining.

````python
# Type your code below
def available_item():
    print('='*20)
    for item in inventory:
        print(item,':',inventory[item])
    print('='*10)

available_item()
````

**For customer to buy an item, we need to write a function for them to call.**



3. Write a function call ```buy(item_name, number_of_item)```, update the dictionary and return a string "OK!"

````python
# Type your code below
def buy(item_name, number_of_item):
    inventory[item_name]-=number_of_item
````

4. Try calling ```buy("Cup", 2)```, then call ```available_item()``` to see if the inventory is updated.

7. Modify the function ```buy(item_name, number_of_item)```, in case the customer buy too many, you return `"Sorry, we only have {number_of_item} {item_name}."`.


````python
# Type your code below
def buy(item_name, number_of_item):
    if item_name in inventory:
        if number_of_item>=inventory[item_name]:
            print(f'Sorry we only have {inventory[item_name]} {item_name}(s)')
        else:
            inventory[item_name]-=number_of_item
    else:
        print("No such item")
````

9. Modify the function buy with English grammer consideration so that ```buy("Fork", 6)``` returns `"Sorry, we only have 5 forks."` or ```buy("Spoon", 6)``` returns `"Sorry, we only have 1 spoon."`

````python
# Type your code below
def buy(item_name, number_of_item):
    if item_name in inventory:
        if number_of_item>=inventory[item_name]:
            if inventory[item_name]<=1: print(f'Sorry we only have {inventory[item_name]} {item_name.lower()}.')
            if inventory[item_name]>1: print(f'Sorry we only have {inventory[item_name]} {item_name.lower()}s.')
        else:
            inventory[item_name]-=number_of_item
    else:
        print("No such item")
        
````
