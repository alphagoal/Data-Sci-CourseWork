### Department Store Data

```python
# Load pandas and the df here

import pandas as pd

store_df = pd.read_csv("https://drive.google.com/u/1/uc?id=1kbCDXONvf8Bn4kEW1ssmxdJLDHC-oV-E&export=download")


store_df.head(10)
```

**Exercise 1**







1. What is the first working and last working week for each Dept-Store?



2. Calculate how many weeks each of them operated.



- Hint1: last working week - first working week + 1



- Hint2: Check the data type of `Date` column

```python
# Your code here
# reference: https://sparkbyexamples.com/pandas/pandas-convert-string-column-to-datetime/
store_df['Date']=pd.to_datetime(store_df['Date'],format="%d/%m/%Y")
df1=store_df[['Store','Dept','Date']].groupby(['Dept','Store']).agg(['min','max'])
```

```python
# print(df1.columns)
# df1.loc[(1,1)]
# print(len(df1.loc[1]))
# print(df1[('Date','min')])

df1['weeks_length']=(df1[('Date','max')]-df1[('Date','min')])\
    .apply(lambda diff: diff.days//7+1)

df1
```
